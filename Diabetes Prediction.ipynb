```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('/content/diabetes.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data['Outcome'].value_counts()
```

0->Non-diabetic
1->Diabetic

```python
X=data.drop(columns='Outcome',axis=1)
Y=data['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Data Standardization
scaler=StandardScaler()
scaler.fit(X)
```

```python
standarized_data=scaler.transform(X)
```

```python
print(standarized_data)
```

```python
X=standarized_data
Y=data['Outcome']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Training the model
classifier=svm.SVC(kernel='linear')
```

```python
#Training data to classifier
classifier.fit(X_train,Y_train)
```

```python
#Model Evaluation
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy score of the training data: ",training_data_accuracy)
```

```python
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy score of the testing data: ",test_data_accuracy)
```

```python
#Making a predictive system
input_data=(6,148,72,35,0,33.6,0.627,50)
#change input data to a numpy array
input_data_as_numpy_array=np.asarray(input_data)
#reshape the array if we dont reshape the array the model expects all the data which is available from us
input_data_reshape=input_data_as_numpy_array.reshape(1,-1)
std_data=scaler.transform(input_data_reshape)
print(std_data)

prediction=classifier.predict(std_data)
print(prediction)
```

```python
if (prediction[0]==0):
  print("The person is non-diabetic")
else:
  print("The person is diabetic")
```
